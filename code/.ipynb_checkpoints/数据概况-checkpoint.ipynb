## 定义一个函数，把复数表达的 i 改为 j

第二次和第三次的数据的复数用 **i** 表示，第一次的是用 **j**，numpy只能识别 **j** 作为复数，所以要把第二次后面的 **j** 替换成 **i**

```python
def i2j(file_path):
    with open(file_path, 'r') as file:
        # Read the content of the file
        content = file.read()

    # Perform the string replacement
    modified_content = content.replace("i", "j")

    # Open the file in write mode to overwrite its content
    with open(file_path, 'w') as file:
        # Write the modified content back to the file
        file.write(modified_content)
```

```python
datasets = [dataset2, dataset3]
for ds in datasets:
    for rom in ds:
        for f in rom:
            i2j(f)
            print(f,'done')
```

## 把 mat 文件转换为 txt 文件

```python
matdir = ['../数据集/dataset2/房间3/真值/']

for md in matdir:
    matfiles = [f for f in os.listdir(md) if f.find('mat') != -1]
    for f in matfiles:
        mat_file = sio.loadmat(md + f)
        txt_file = md + f.replace('mat', 'txt')
        print(f'{f}', end=' ')
#         print(mat_file['truth'])
        np.savetxt(txt_file, mat_file['truth'], fmt='%d', delimiter='\t')
                
        print(f'converted to txt')
```

## 把数据数据 reshape 后存为 npy 的格式

### dataset1 to npy

```python
for rom in dataset1:
    for f in rom:
        data = np.loadtxt(f, dtype=complex)
        print(data.shape, end='')
        data = data.reshape((-1, head_len +64*4))
        
        dst = f.replace('txt', 'npy')
        print('->', data.shape, dst)
        np.save(dst, data)
```
